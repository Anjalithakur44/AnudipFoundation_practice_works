```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

db_config = {
    'user': 'root',
    'password': '@anjali123%',
    'host': 'localhost',
    'database': 'ecommerce_db'
}

def get_connection():
    """Establish a connection to the MySQL database."""
    return mysql.connector.connect(**db_config)

def fetch_data(query):
    """Fetch data from the database and return as a Pandas DataFrame."""
    conn = get_connection()
    df = pd.read_sql(query, conn)
    conn.close()
    return df

def clean_data(df):
    """Clean the data by handling missing values and duplicates."""
    if 'order_date' in df.columns and 'order_status' in df.columns:
        df.fillna({
            'order_date': pd.Timestamp.now(),  
            'order_status': 'Pending'  
        }, inplace=True)
        df.drop_duplicates(inplace=True)
        df['order_date'] = pd.to_datetime(df['order_date'])
    return df

def plot_sales_over_time(df):
    """Plot sales over time."""
    if 'order_date' in df.columns and 'total_price' in df.columns:
        df['total_price'] = pd.to_numeric(df['total_price'], errors='coerce')
        sales_over_time = df.groupby(df['order_date'].dt.date)['total_price'].sum()
        plt.figure(figsize=(12, 6))
        plt.plot(sales_over_time.index, sales_over_time.values, marker='o')
        plt.title('Sales Over Time')
        plt.xlabel('Date')
        plt.ylabel('Total Sales')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

def plot_total_sales_by_product(df, products_df):
    """Plot total sales by product."""
    if 'product_id' in df.columns and 'total_price' in df.columns:
        df['total_price'] = pd.to_numeric(df['total_price'], errors='coerce')
        product_sales = df.groupby('product_id').agg({'total_price': 'sum'}).reset_index()
        product_sales = product_sales.merge(products_df[['product_id', 'product_name']], on='product_id')
        plt.figure(figsize=(12, 6))
        plt.bar(product_sales['product_name'], product_sales['total_price'], color="red")
        plt.title('Total Sales by Product')
        plt.xlabel('Product')
        plt.ylabel('Total Sales')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

def plot_payment_mode_distribution(df):
    """Plot the distribution of payment modes."""
    if 'payment_mode' in df.columns:
        payment_mode_distribution = df['payment_mode'].value_counts()
        plt.figure(figsize=(12, 6))
        plt.bar(payment_mode_distribution.index, payment_mode_distribution.values, color="orange")
        plt.title('Distribution of Payment Modes')
        plt.xlabel('Payment Mode')
        plt.ylabel('Number of Orders')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

def add_new_customer(data):
    """Add a new customer to the customer table."""
    conn = get_connection()
    cursor = conn.cursor()

    query = """
    INSERT INTO customer (customer_id, name, city, email, phone_no, address, pin_code)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(query, (
        data['customer_id'], data['name'], data['city'], data['email'],
        data['phone_no'], data['address'], data['pin_code']
    ))
    conn.commit()

    cursor.close()
    conn.close()

def add_new_product(data):
    """Add a new product to the product table."""
    conn = get_connection()
    cursor = conn.cursor()

    query = """
    INSERT INTO product (product_id, product_name, category, sub_category, original_price, selling_price, stock)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(query, (
        data['product_id'], data['product_name'], data['category'],
        data['sub_category'], data['original_price'], data['selling_price'], data['stock']
    ))
    conn.commit()

    cursor.close()
    conn.close()

def add_new_order(data):
    """Add a new order to the order_details table."""
    conn = get_connection()
    cursor = conn.cursor()

    query = """
    INSERT INTO order_details (order_id, customer_id, product_id, quantity, total_price, payment_mode, order_date, order_status)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(query, (
        data['order_id'], data['customer_id'], data['product_id'], data['quantity'],
        data['total_price'], data['payment_mode'], data['order_date'], data['order_status']
    ))
    conn.commit()

    cursor.close()
    conn.close()

def main():
    """Main function to provide an interactive command-line interface."""
    while True:
        print("Select an option:")
        print("1. View Customers")
        print("2. View Products")
        print("3. View Orders")
        print("4. Data Cleaning")
        print("5. Data Visualization")
        print("6. Add New Customer")
        print("7. Add New Product")
        print("8. Add New Order")
        print("9. Exit")

        option = input("Enter option number: ")

        if option == '1':
            df = fetch_data('SELECT * FROM customer')
            print(df)
        elif option == '2':
            df = fetch_data('SELECT * FROM product')
            print(df)
        elif option == '3':
            df = fetch_data('SELECT * FROM order_details')
            cleaned_df = clean_data(df)
            print(cleaned_df)
        elif option == '4':
            df = fetch_data('SELECT * FROM order_details')
            cleaned_df = clean_data(df)
            print("Data cleaning completed. Here is the cleaned data:")
            print(cleaned_df)
        elif option == '5':
            df = fetch_data('SELECT * FROM order_details')
            cleaned_df = clean_data(df)
            while True:
                print("\nSelect a visualization option:")
                print("1. Sales Over Time")
                print("2. Total Sales by Product")
                print("3. Payment Mode Distribution")
                print("4. Back to Main Menu")

                vis_option = input("Enter option number: ")

                if vis_option == '1':
                    plot_sales_over_time(cleaned_df)
                elif vis_option == '2':
                    products_df = fetch_data('SELECT * FROM product')
                    plot_total_sales_by_product(cleaned_df, products_df)
                elif vis_option == '3':
                    plot_payment_mode_distribution(cleaned_df)
                elif vis_option == '4':
                    break
                else:
                    print("Invalid option. Please try again.")
        elif option == '6':
            data = {
                'customer_id': input("Enter customer ID: "),
                'name': input("Enter customer name: "),
                'city': input("Enter city: "),
                'email': input("Enter email: "),
                'phone_no': input("Enter phone number: "),
                'address': input("Enter address: "),
                'pin_code': input("Enter pin code: ")
            }
            add_new_customer(data)
            print("New customer added successfully.")
        elif option == '7':
            data = {
                'product_id': input("Enter product ID: "),
                'product_name': input("Enter product name: "),
                'category': input("Enter category: "),
                'sub_category': input("Enter sub-category: "),
                'original_price': input("Enter original price: "),
                'selling_price': input("Enter selling price: "),
                'stock': input("Enter stock: ")
            }
            add_new_product(data)
            print("New product added successfully.")
        elif option == '8':
            data = {
                'order_id': input("Enter order ID: "),
                'customer_id': input("Enter customer ID: "),
                'product_id': input("Enter product ID: "),
                'quantity': input("Enter quantity: "),
                'total_price': input("Enter total price: "),
                'payment_mode': input("Enter payment mode: "),
                'order_date': input("Enter order date (YYYY-MM-DD): "),
                'order_status': input("Enter order status: ")
            }
            add_new_order(data)
            print("New order added successfully.")
        elif option == '9':
            break
        else:
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    main()
```
